```python
import pandas as pd


file_path = 'monthly_sales_data.csv'  
sales_data = pd.read_csv(file_path)

print(sales_data.head())
```

```python
sales_data['Date'] = pd.to_datetime(sales_data['Date'], dayfirst=True)

sales_data['Quantity Sold'] = pd.to_numeric(sales_data['Quantity Sold'], errors='coerce')
sales_data['Sales Amount'] = pd.to_numeric(sales_data['Sales Amount'], errors='coerce')
```

```python
total_sales_per_product = sales_data.groupby('Product')['Sales Amount'].sum().reset_index()

total_sales_per_product.sort_values(by='Sales Amount', ascending=False, inplace=True)

print(total_sales_per_product)
```

```python

sales_data['Month'] = sales_data['Date'].dt.to_period('M')

monthly_sales_trends = sales_data.groupby('Month')['Sales Amount'].sum().reset_index()

print(monthly_sales_trends)
```

```python

top_selling_product = total_sales_per_product.iloc[0]['Product']
print(f"The top-selling product is: {top_selling_product}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(x='Sales Amount', y='Product', data=total_sales_per_product, hue='Product', palette='Blues_d', legend=False)
plt.title('Total Sales per Product')
plt.xlabel('Sales Amount')
plt.ylabel('Product')
plt.show()

if isinstance(monthly_sales_trends['Month'].dtype, pd.PeriodDtype):
    monthly_sales_trends['Month'] = monthly_sales_trends['Month'].dt.to_timestamp()

monthly_sales_trends['Month'] = pd.to_datetime(monthly_sales_trends['Month'])

plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Sales Amount', data=monthly_sales_trends, marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45)
plt.show()
```
